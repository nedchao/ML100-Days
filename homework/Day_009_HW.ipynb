### Correlation Coefficient
#### 以下程式碼將示範在 python 如何利用 numpy 計算出兩組數據之間的相關係數，並觀察散佈圖

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

np.random.seed(1) 
```

### 正相關 Positive correlation 

```python
#隨機生成 1000 個介於 0~50 的數
x = np.random.randint(0, 50, 1000)

# 正相關，增加一些雜訊
y = x + np.random.normal(0, 10, 1000)

# 可用 numpy 裡寫好的 function 來計算相關係數
np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

### 負相關 Negative correlation 

```python
# 隨機生成 1000 個介於 0~50 的數
x = np.random.randint(0, 50, 1000)

# 負相關，增加一些雜訊
y = 100 - x + np.random.normal(0, 5, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

### 弱相關 No correlation

```python
x = np.random.randint(0, 50, 1000)
y = np.random.randint(0, 50, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```
